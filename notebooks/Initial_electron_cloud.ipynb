## Source: generation of electron cloud 


```python
# Create an instance of the Source class
source = Source('../data/Ba133_expanded5k.csv')

# Read the energies from the file
source.read_energies()
```

```python
# Choose a specific interaction energy and electric field ADD UNITS
interaction = nestpy.nr
density = 2.9 
field = 100.

n_electrons = source.n_initial_electrons(interaction, field)
```

```python
print(n_electrons)
```

```python
# Initial positions of electrons
positions = source.pos_initial_electrons(n_electrons)
```

```python
unique_positions = np.unique(positions, axis=0) 
#print(unique_positions)
```

```python
# Extract x and y positions from the positions array
x_positions = unique_positions[:, 0] 
y_positions = unique_positions[:, 1]  # add +25 if file from G4 because TPC is not centered in G4 simulation

# Truncate n_electrons array to match the length of x_positions and y_positions
n_electrons_truncated = n_electrons[:len(x_positions)]

# Plot positions 
pu.scatter_plot(x_positions, y_positions, n_electrons_truncated, 'Ba133', 'Ba133')
```
